```python
import json, pandas as pd
```

```python
f = open('test.json', 'r')

test_data = json.loads(f.read())
filter_test = [x for x in test_data['data'] if x['task'] == 'textual_cloze']

output_json = json.dumps(filter_test, indent = 4)

with open('test_filtered.json', 'w') as outfile:
    outfile.write(output_json)
```

```python
#combine body and delete unnecessary attributes: videos, images, id
file_name = ['train', 'test', 'val']
for file in file_name:
    #combine body context for filtered test file
    json_file = open('{file}_filtered.json'.format(file=file), 'r')
    open_json = json.loads(json_file.read())

    for i in range(len(open_json)):
        context = ""
        for j in range(len(open_json[i]['context'])):
            context += open_json[i]['context'][j]['body']
            del open_json[i]['context'][j]['videos']
            del open_json[i]['context'][j]['images']
            del open_json[i]['context'][j]['id']
            del open_json[i]['context'][j]['body']
        open_json[i]['context'] = context
        
    output = json.dumps(open_json, indent = 4)
    with open('combine_body_{file}.json'.format(file=file), 'w') as outfile:
        outfile.write(output)
```
